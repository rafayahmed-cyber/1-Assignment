### 43. Ek simple calculator banayein (+, -, *, /) symbols ke basis par operation perform karein.

```python
num1 = input("Enter first number: ")
num2 = input("Enter second number: ")
op = input("Enter operation (+, -, *, /): ")

num1 = float(num1)
num2 = float(num2)

if op == "+":
    print(num1 + num2)
elif op == "-":
    print(num1 - num2)
elif op == "*":
    print(num1 * num2)
elif op == "/":
    print(num1 / num2)
else:
    print("Invalid operation")

```

### 44. User ka weight aur height le kar BMI category batayein (Underweight, Normal, Overweight).

```python
weight = input("Enter weight (kg): ")
height = input("Enter height (meters): ")

weight = float(weight)
height = float(height)

bmi = weight / (height * height)

if bmi < 18.5:
    print("Underweight")
elif bmi < 25:
    print("Normal")
else:
    print("Overweight")
```

### 58. Electricity Bill: Pehle 100 units free, next 100 units Rs 5/unit, uske baad Rs 10/unit.

```python
units = input("Enter electricity units: ")
units = int(units)

bill = 0

if units <= 100:
    bill = 0
elif units <= 200:
    bill = (units - 100) * 5
else:
    bill = (100 * 5) + (units - 200) * 10

print("Total Bill:", bill)
```
